```python
import pandas as pd

# Read the first dataset
df1 = pd.read_csv('full_merged_dataset14.csv', encoding='utf-8', low_memory=False)
```

```python
# Read the second dataset
df2 = pd.read_csv('full_poems_data.csv', encoding='utf-8')
```

```python
# Merge the datasets on a shared column
combined_df = pd.concat([df1, df2], ignore_index=True)

# Save the merged dataset to a new CSV file
combined_df.to_csv('full_merged_dataset15.csv', index=False) # CAREFUL, you can overwrite your previous dataset, UPDATE the name when used
```
